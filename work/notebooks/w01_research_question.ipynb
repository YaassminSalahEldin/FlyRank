## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

**Observation 1:** The near-zero correlation ($0.001$) proves that high keyword search volume does not guarantee actual page impressions.

**Observation 2:** Declining (down) pages actually have a lower median age ($216$ days) compared to growing (up) pages ($292$ days), proving that traffic decay is not just a function of sheer "page age".Paste this into the CODE cell below the text:

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import pandas as pd

url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Number 1: Search Volume vs Impressions
active_df = df[df["impressions_90d"] > 0]
corr_active = active_df["search_volume"].corr(active_df["impressions_90d"])
print(f"Active Correlation: {corr_active:.3f}")

# Number 2: Content Age vs Trend
age_trend = df.groupby("trend_direction")["content_age_days"].median()
print("\nMedian Content Age by Trend:")
print(age_trend.to_string())
```
